**Project Aim**: Video Stream Ingestion

Simulate the ingestion of a live video stream using Python using libraries like
OpenCV to read video frames from a video file. Consider that as a live stream and
provide code to capture frames continuously from the source and store in separate image files.

 **Solution Approach:** Use OpenCV to capture and process frame wise data from a captured video.
 Opencv is an open source library which is very useful for computer vision applications such as video analysis, CCTV footage analysis and image analysis.
 for more detailed study refer site:

```python
# Program To Read video
# and Extract Frames

import cv2

timestamps = []

# Function to extract frames
def FrameCapture(path):
  # Path to video file
  vidObj = cv2.VideoCapture(path)

  # Used as counter variable
  count = 0

  #checks whether frames were extracted
  success = 1

  # find end of frames count
  end = int(vidObj.get(cv2.CAP_PROP_FRAME_COUNT))

  if vidObj.isOpened():

    while (count < end) :
      # vidObj object calls read
      # function extract frames
      success, image = vidObj.read()

      # Saves the frames with frame-count
      cv2.imwrite("frame%d.jpg" % count, image)

      timestamps.append(vidObj.get(cv2.CAP_PROP_POS_MSEC))

      count += 1


# Driver code
url = "https://www.youtube.com/watch?v=1InvZ4TF3Hg"
#pth = "C:\\Users\\janani\\Downloads\\Happy monks.mp4"
p1 = "Happy monks.mp4"
FrameCapture(p1)

```

```python
from PIL import Image
from PIL.ExifTags import TAGS

# open the image
image = Image.open("img.jpg")

# extracting the exif metadata
exifdata = image.getexif()

# looping through all the tags present in exifdata
for tagid in exifdata:
  # getting the tag name instead of tag id
  tagname = TAGS.get(tagid, tagid)

  # passing the tagid to get its respective value
  value = exifdata.get(tagid)

  # printing the final result
  print(f"{tagname:25}: {value}")
```
