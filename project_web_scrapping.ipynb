## Python for DS - Web Scrapping Project

### Step 1: Import required libraries for establishing link to specific websites and download data from the sites

```python
# import request library which is used to query/obtain info from a website
import requests
# specify the url
web_link = "https://en.wikipedia.org/wiki/List_of_island_countries"
# fetching links from the webpage
link = requests.get(web_link).text
```

```python
# to check whether links are fetched properly, print and verify
print(link)
```

### Step 2: Parse the data returned from the website

```python
# parse the html and link variable using beautiful soup
from bs4 import BeautifulSoup
soup = BeautifulSoup(link, 'lxml')
#lxml is a Python library which allows for easy handling of XML and HTML files, and can also be used for web scraping
print(soup)
```

```python
# since data is still not that readable using prettify to make it more readable i.e it gets arranged into nested loops structure
print(soup.prettify())
```

### Step 3: Working with html tags

```python
# print title of webpage
soup.title
```

```python
# to remove tags separate string part from the result
print("Title of webpage: {}".format(soup.title.string))
```

```python
# to analyse links find all tags which has a tag in it
soup.a
```

```python
# fetched only one result. so inorder to find all tags containing specific parameters we use findall function
soup.find_all('a')
```

```python
# to display all the links without tags, iterate over the entire a tags list and get string starting with href
all_link = soup.find_all('a')
for link in all_link:
    print(link.get("href"))
```

```python
# fetch all table tags
all_table = soup.find_all("table")
print(all_table)
```

### Step 4: Identify the right table - use attribute of class of table and use it to filter the right table class

```python
# to get data from a particular table locate it and get its class name by right clicking on it and inspect
# use the class name to sort through table tags
user_table = soup.find('table', class_ = "wikitable sortable")
user_table
```

```python
# actual data that we want i.e list of  island nations under UN, that data is available in all 'a' tags
# so use findall function and extract that particukar data and store in a variable
table_link = user_table.findAll('a')
table_link
```

```python
# to print without tags, iterate over the entire link list and get strings starting with href
for link in table_link:
    print(link.get('title'))
```

### Step 5: Import the data extracted/ Scrapped into a datframe

```python
import pandas as pd
# define a dataframe
df = pd.DataFrame(columns=['Country','ISO code','Population','Density'])

# collecting data
for row in user_table.tbody.find_all('tr'):
    # find all data for each column
    columns = row.findAll('td')

    if(columns != []):
        country = columns[0].text.strip()
        iso_code = columns[1].text.strip()
        population = columns[5]
        density = columns[6]

        df = df.append({'Country': country, 'ISO code': iso_code, 'Population': population, 'Density': density}, ignore_index=True)
```
